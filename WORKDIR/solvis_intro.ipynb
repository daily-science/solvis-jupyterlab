# Using `solvis` with NSHM Composite Solution



The file `NSHM_v1.0.4_CompositeSolution.zip` is an archive file that contains aggregated dataset for the three Fault Systems that make up the NSHM:
 -  **CRU** = the NZ crustal fault system, with 36 logic tree branches 
 -  **HIK** = the Hikurangi-Kermadec subduction fault system, with 9 logic tree branches
 -  **PUY** = the Puysegur subduction fault, with 3

Each logic tree branch produced an opensha **InversionSolution** archive. The **solvis** library provides python tools for inspecting both these files, and  
also the aggregated **CompositeSolution** archive.

**NB** there is one additional Source model (The distributed source model) that is not handled with solvis.  

### Load the NSHM composite solution

```python
import json
from IPython.display import GeoJSON

# NSHM libraries ..
import nzshm_model as nm
import solvis

# we need a matching source_logic_tree (CBC do we really ???)
slt = nm.get_model_version("NSHM_v1.0.4").source_logic_tree

# load the composite_solution using the source logic tree
csol = solvis.CompositeSolution.from_archive("NSHM_v1.0.4_CompositeSolution.zip", slt)

# get the the Crustal FaultSystemSolution
fss = csol._solutions['CRU']
```

## Display the three fault systems (CRU, PUY, HIK)

Each fault system solution contains a geojson dataframe with all the geometries and other scientific attributes used to produce the Inversion Solutions.

```python
fss = csol._solutions['CRU']
gj = json.loads(fss.fault_surfaces().to_json())
GeoJSON(gj) # to view the surfaces in a cell....
```

```python
# now Puysegur
fss = csol._solutions['PUY']
gj = json.loads(fss.fault_surfaces().to_json())
GeoJSON(gj)
```

```python
# and Hikurangi
fss = csol._solutions['HIK']
gj = json.loads(fss.fault_surfaces().to_json())
GeoJSON(gj)
```
